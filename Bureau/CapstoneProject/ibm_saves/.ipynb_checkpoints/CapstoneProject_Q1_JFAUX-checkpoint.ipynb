````python
print(df_data_1.complaint_type.value_counts(), len(df_data_1.complaint_type.value_counts()))
````

### Answer:

````python
heat_complaints = df_data_1.complaint_type.value_counts().values[0]+df_data_1.complaint_type.value_counts().values[1]
all_complaints = df_data_1.complaint_type.value_counts().values.sum()
heat_complaint_rate = str(100*heat_complaints/all_complaints)[:2] + " %"
print("The DHPP of NYC shall focus on heat / hot water / heating complaints first, because we count 30 different complaint types and heat / hot water / heating represent more than %s of all complaints." % heat_complaint_rate)
````
