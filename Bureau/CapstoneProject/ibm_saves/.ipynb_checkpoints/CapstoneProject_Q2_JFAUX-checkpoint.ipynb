# Data Preparing
## First clean

We will now add a specific 'is_heat_the_problem' column using condition on complaint type to detect correlation between complaint type and other features.

Assuming heat/hot water & heating have same root cause (boiler), we can regroup them.

```python
# reduce memory usage by droping columns unused in this notebook (borough will be used from pluto df)
df_data_2 = df_data_2[["complaint_type","incident_zip","incident_address","street_name", "latitude", "longitude"]]
df_data_2 = df_data_2.dropna()
```

```python
# make further dataframes treatments more readable by adding a boolean "is_heat_the_problem"
# 'is_heat_the_problem' will become our Å¶ in the 4th chapter
df_data_2['is_heat_the_problem'] = np.where((df_data_2['complaint_type']=='HEAT/HOT WATER') | (df_data_2['complaint_type']=='HEATING'), True, False)
```

```python
#data preparation : we set addresses spaces to 1 for better matching :) it optimized the merge by adding + 488K rows
df_data_2['incident_address'] = [' '.join(str(mystring).split()) for mystring in df_data_2['incident_address']]
df_data_3.address = [' '.join(str(mystring).split()) for mystring in df_data_3.address]
```

```python
df_data_3 = df_data_3[['borough', 'schooldist', 'address', 'numfloors', 'yearbuilt']]
df_data_3 = df_data_3.dropna()
```

# dataframes merging

Now is the time to merge pluto & NYC complaints

```python
df_data_2 = pd.merge(df_data_2, df_data_3, how='left', left_on=['incident_address'], right_on = ['address'])
```

```python
df_data_2 = df_data_2.dropna()
df_data_2['incident_zip'] = df_data_2['incident_zip'].astype('int16')
df_data_2['numfloors'] = df_data_2['numfloors'].astype("int8")
df_data_2['schooldist'] = df_data_2['schooldist'].astype("int16")
df_data_2['yearbuilt'] = df_data_2['yearbuilt'].astype('int16')
df_data_2.head()
```

Now that data is ready for analysis, let's first have a look at the specific distrubtion of heat/hot water/heating complaints on NYC's map

### NYC Complaints Map visualization

In order to get an idea of where the heating problems occur most, let's plot them !
Since dataframe is pretty big, we will plot 0,1 percent of overall complaints by passing frac=.001 to sample method :

```python
df = df_data_2[["complaint_type", "latitude", "longitude"]].sample(frac=.001)
nyc_map = folium.Map(location=[df.latitude.mean(), df.longitude.mean()], zoom_start=11)
incidents = folium.map.FeatureGroup()
incidents = plugins.MarkerCluster().add_to(nyc_map)
for inc, lat, lng, in zip(df.complaint_type, df.latitude, df.longitude):
    if inc == 'HEAT/HOT WATER' or inc == 'HEATING':
        folium.Marker([lat, lng],icon=None).add_to(incidents)
nyc_map
```

*(You may doubleclic in the blank zone on the left to hide map)*

Even if dots are randomly picked, above map shows non linear distribution of heat complaints, with **more density** in **Brooklyn, Bronx & Manhattan** boroughs.

# Complaints analysis per borough

## Preamble

```python
# focus only on heat complaints
df_heat = df_data_2.loc[df_data_2['is_heat_the_problem'] == True]
# hcn = heat complaint number
df_heat = df_heat[['complaint_type', 'incident_zip', 'incident_address', 'street_name','borough', 'latitude', 'longitude']]
total_hcn = df_heat.shape[0]
boroughs_hc_values=[]
for x in range(0, len(df_heat['borough'].value_counts())):
    #print(df_heat['borough'].value_counts().index[x], df_heat['borough'].value_counts()[x])
    boroughs_hc_values.append([df_heat['borough'].value_counts().index[x], df_heat['borough'].value_counts()[x]])
    
import matplotlib.pyplot as plt
%matplotlib inline
labels, sizes = [x[0] for x in boroughs_hc_values], [100 * x[1]/ total_hcn for x in boroughs_hc_values]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', "turquoise", "orange"]
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=False, startangle=0)
plt.title("NYC HEAT COMPLAINT REPARTITION PER BOROUGH")
plt.axis('equal')
plt.show()
```

above pie shows :
    - heterogenous repartition of heat compaints per borough;
    - some boroughs might have a predisposition to heat complaints : Bronx, Manhattan & Brooklyn;
    - heat complaint prediction could be harder for boroughs like Staten Island & Queens;

```python
# counting all complaints per borough to get relative proportion of heat complaint rate within a borough
# note that is this loop , full data2 is used instead of specific df_heat
boroughs_hc_values=[]
for x in range(0, len(df_heat['borough'].value_counts())):
    boroughs_hc_values.append([df_heat['borough'].value_counts().index[x], df_heat['borough'].value_counts()[x]])
print("boroughs heat complaints values OK")
boroughs_tc_values  = []
for x in range(0, len(df_data_2['borough'].value_counts())):
    boroughs_tc_values.append([df_data_2['borough'].value_counts().index[x], df_data_2['borough'].value_counts()[x]])
print("total complaint values per boroughs OK")
borough_hc_rates = []
labels = [x[0] for x in boroughs_tc_values]
for label in labels:
    for hc_items in boroughs_hc_values:
        for tc_items in boroughs_tc_values:
            if label == hc_items[0] and label == tc_items[0]:
                borough_hc_rates.append((label, 100* hc_items[1]/tc_items[1]))
print("borough_hc_rates OK")
```

```python
from gc import collect # optimize exec on remote notebook , #df_data_2
del [[df_data_3]]
print(collect())
```

```python
labels, values = [x[0] for x in borough_hc_rates], [x[1] for x in borough_hc_rates]
x=np.arange(len(labels))
plt.bar(x, height=values, color=colors)
labels=tuple(labels)
plt.gca().set_ylim(20, 40)
plt.xticks(x, labels,  rotation=45)
plt.ylabel('percentage %')
plt.title("HEAT COMPLAINT PROPORTION PER BOROUGH")
```

Above bar chart confirms Staten Island has the less heat complaint proportion.
Above bar chart shows **Mahattan, Queens & Bronx perfect targets for complaint prediction per borough**.

# Complaints analysis per ZipCode

```python
df_heat['incident_zip'] = df_heat['incident_zip'].astype('str')
df_data_2['incident_zip'] = df_data_2['incident_zip'].astype('str')
#df_heat['incident_zip'].value_counts()[0]
```

```python
print(df_data_2[df_data_2['incident_zip'] == '10045']['borough'].values[0])
```

```python
zip_hc_values=[]
for x in range(0, len(df_heat['incident_zip'].value_counts())):
    zip_hc_values.append((df_heat['incident_zip'].value_counts().index[x], df_heat['incident_zip'].value_counts()[x]))
zip_tc_values  = []
print("heat complaint values per zip OK, len :", len(zip_hc_values))
for x in range(0, len(df_data_2['incident_zip'].value_counts())):
    zip_tc_values.append((df_data_2['incident_zip'].value_counts().index[x], df_data_2['incident_zip'].value_counts()[x]))
print("total complaint values per zip OK, len :", len(zip_tc_values))
zip_hc_rates = []
labels = [x[0] for x in zip_tc_values]
i=0
for label in labels:
    for hc_items in zip_hc_values:
        for tc_items in zip_tc_values:
            if label == hc_items[0] and label == tc_items[0]:
                zip_hc_rates.append((label, 100* hc_items[1]/tc_items[1], df_data_2[df_data_2['incident_zip'] == label]['borough'].values[0]))
print("zip_hc_rates OK")
```

```python
# sort zipcodes in order of appearence desc
sorted_by_second = sorted(zip_hc_rates, key=lambda tup: tup[1])
# append color in color list & pass it to barh
labels, values = [x[0] for x in sorted_by_second], [x[1] for x in sorted_by_second]
borough_color_map={'MN':'yellowgreen', 
                  'BK':'gold',
                  'BX': 'lightskyblue',
                  'SI': 'lightcoral',
                  'QN': "turquoise"}
```

```python
x=np.arange(len(labels))
my_barh = plt.barh(x, width=values)
for elem in range(len(my_barh)):
    my_barh[elem].set_color(borough_color_map[sorted_by_second[elem][2]])
labels=tuple(labels)
plt.rcParams["figure.figsize"] = (50, 50)
#plt.gca().set_ylim(0, 120)
plt.yticks(x, labels,  rotation=45)
plt.xlabel('percentage %')
plt.title("HEAT COMPLAINT PROPORTION PER ZIP")
```

bar chart above shows several interesting elements :
    - some complaints per zipcode are exclusively related to heat problems;
    - 3 of the 5 zipcodes exclusively related to heat problems come from Manhattan (yellowgreen color);
    - Staten Island zipcodes have low heat complaint rate;
    - manhattan zipcodes have high complaint rate;

# Complaint analysis at Street level 

```python
most_crapy_streets = []
podium_size = 50
for i in range(podium_size):
    most_crapy_streets.append(df_data_2['street_name'].value_counts().index[i])
```

```python
most_crapy_borough_by_street_analysis = []
for label in most_crapy_streets:
    most_crapy_borough_by_street_analysis.append(df_data_2[df_data_2['street_name'] == label]['borough'].values[0])
result = [(x, most_crapy_borough_by_street_analysis.count(x)) for x in set(most_crapy_borough_by_street_analysis)]
print(result)
```

We can see that :
    - Queens & Staten Island boroughs do not figure in the top 50 of the streets with most complaints
    - Brooklyn is the winner with 21 streets out of 50

Below is the plotting data & chart of top 50 complaints seen at the street level with a specific color for each borough

```python
# /!\ took ~ 15 min to compute
most_crapy_streets = []
podium_size = 50
for i in range(podium_size):
    most_crapy_streets.append((df_data_2['street_name'].value_counts().index[i], df_data_2['street_name'].value_counts()[i], df_data_2[df_data_2['street_name'] == df_data_2['street_name'].value_counts().index[i]]['borough'].values[0]))
```

```python
# sort streets in order of appearence desc
sorted_by_second = sorted(most_crapy_streets, key=lambda tup: tup[1])
# append color in color list & pass it to barh
labels, values = [x[0] for x in most_crapy_streets], [x[1] for x in most_crapy_streets]
labels.reverse()
values.reverse()
borough_color_map={'MN':'yellowgreen', 
                  'BK':'gold',
                  'BX': 'lightskyblue',
                  'SI': 'lightcoral',
                  'QN': "turquoise"}
x=np.arange(len(labels))
my_barh = plt.barh(x, width=values)
for elem in range(len(my_barh)):
    my_barh[elem].set_color(borough_color_map[sorted_by_second[elem][2]])
for i, v in enumerate(values):
    plt.text(v + 10, i + 0, str(v), size=20)
labels=tuple(labels)
plt.rcParams["figure.figsize"] = (40, 40)
#plt.gca().set_ylim(0, 120)
plt.yticks(x, labels,  rotation=45)
plt.xlabel('complaints number')
plt.ylabel('street name')
plt.title("TOP STREET LEVEL HEAT COMPLAINTS & THEIR ASSOCIATED BOROUGH")
```

to conclude this notebook, it seems that prediction will be easier on Manhattan, Bronx & Brooklyn boroughs, whatever zipcode or street name having enough entries to train & test model.
